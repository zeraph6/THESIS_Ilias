```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

from matplotlib.ticker import ScalarFormatter

# Data Preparation
sizes = ['25GB', '100GB', '1B']
methods = ['L-NN', 'Vnode-set']

x = np.arange(len(methods))

method_colors = ['#1f77b4', '#ff7f0e']  # Colors for L, AL2, AL4
dataset_hatches = ['//', 'xx']  # Patterns for Deep and SIFT
```

```python

# Example data (using random values for illustration)
deep_times = { 32, 40
    '25GB': [7600 / 3600, 84000 / 3600],
    '100GB': [28500 / 3600, 33656 / 3600],
    '1B': [116216 / 3600, 91954 / 3600]
}

sift_times = {40 , 49
    '25GB': [7200 / 3600, 6100 / 3600],
    '100GB': [22571 / 3600, 27028 / 3600],
    '1B': [141737 / 3600, 145179 / 3600]
}

bar_width = 0.4

for size in sizes:
    fig, ax = plt.subplots(figsize=(6, 5))
    
    # Plot bars for 'Deep' dataset
    ax.bar(x - bar_width/2, [deep_times[size][j] for j in range(len(methods))], 
           width=bar_width, label='Deep', color='lightblue', hatch=dataset_hatches[0], edgecolor='black')
    
    # Plot bars for 'SIFT' dataset
    ax.bar(x + bar_width/2, [sift_times[size][j] for j in range(len(methods))], 
           width=bar_width, label='SIFT', color='lightgreen', hatch=dataset_hatches[1], edgecolor='black')

    ax.set_xlabel('', fontsize=20)  # Remove x-axis title
    ax.set_xticks(x)
    ax.set_xticklabels(methods, fontsize=20)
    
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.tick_params(axis='both', which='major', labelsize=30)
    
    ax.tick_params(axis='y', which='major', labelsize=25)

    ax.grid(axis='y', linestyle='--')

    if size == '25GB':
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles, labels, fontsize=16, 
                  title_fontsize='14', loc='upper right')
        #ax.set_ylabel('Indexing Time (Hrs)', fontsize=17)

    plt.tight_layout()
    plt.savefig(size+"_Idx.jpg")
    plt.show()
```
